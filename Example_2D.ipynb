## 2-D Problem: Friction Factor of Flow in Rough Pipes

The flow in rough pipes is a classical fluid mechanics problem where the friction factor (f) is a function of the Reynolds number (Re) and the relative roughness (K = roughness / diameter).

### 1. Import the taining dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# import the training dataset (should be saved in .csv)
# the csv file should include a header (1st row) with the names of the input and output variables
df_train = pd.read_csv("data/data_rough.csv")
SIZE = len(df_train)

col_x = ['Re','K'] # list of the names of the input variables
col_y = ['f'] # list of the names of the output variables

# you may visualize the training data
# this is possible for 1D and 2D problem, but may be difficult for multidimensional data 

from matplotlib.colors import ListedColormap
clm=plt.get_cmap('viridis').colors[::1]
clm=ListedColormap(clm)

fig, ax = plt.subplots()
sc = ax.scatter(df_train.Re, df_train.f, s=0.1, c=np.log10(df_train.K), cmap=clm)  
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Re')
ax.set_ylabel('f')

clb = plt.colorbar(sc, ax=ax)
clb.set_label('log₁₀(K)')

plt.show()
```

### 2. Train DimNet models
It may take several minutes to hours to run this cell, depending on the size of the dataset, the size of the network, and more importantly, the number of the repeats (N_REPEAT)

```python
from core import train_DimNet
from datetime import datetime
import os
from joblib import dump
import torch

# config the widths of the two hidden layers of the DimNet, respectively
N_HIDDEN = [2,2]

# set the number of repeats for training DimNets 
# note: each repeat uses a different seed for initialization
N_REPEAT = 20

# set the starting number of the current run of repeats; default = 0
# As you may run this cell for several times, every time you run this cell, 
# set N_START = 1 + # of the last repeat in the last run, so that the previous output files won't be overriden
N_START = 0

# Create a name for the job. 
# The output results, models, etc. will be saved under a folder with this name
NAME = "rough"

mape_ = []
init_seed_ = []

index_ = list(range(N_START,N_START+N_REPEAT))

for i in index_:
    print(f"Repeat #{i}")
    print(f"Training model for hidden-layer config of {N_HIDDEN} with {SIZE} data samples")  
    _,mape,init_seed = train_DimNet(df_train,col_x,col_y,NAME,N_HIDDEN,i_repeat=i,
                                        BATCH_SIZE = len(df_train),
                                        LR = 0.001,
                                        MIN_EPOCH = 20000,
                                        MAX_EPOCH = 100000,
                                        INTERVAL_PRINT = 100,
                                        INTERVAL_SAVE = 100,
                                        WIN_LEN = 100,
                                        CHANGE_RATE_TOLERANCE = 1e-6,
                                        SOLVER = 'Adam',
                                        SHUFFLE = True,
                                        VERBOSITY = 0)
    mape_.append(mape)
    init_seed_.append(init_seed)

sorted_list = sorted(zip(index_,mape_,init_seed_), key=lambda x:x[1])
i_best = sorted_list[0][0]
mape_best = sorted_list[0][1]
```

### 3. Print & log training information

```python
path_log = os.path.join(NAME,'log.txt')
with open(path_log, 'a') as f_log:
    f_log.write('\n\n')
    f_log.write('='*10 + f"{datetime.now()}" + '='*10)
    f_log.write(f'\nsize of training set = {SIZE}\n'+
                f'hidden layer config = {str(N_HIDDEN)}\n'+
                'solver = Adams\n'+
                '\nREPEAT#  MAPE    SEED\n')
    
    print('='*10 + f"{datetime.now()}" + '='*10)
    print(f'\nsize of training set = {SIZE}\n'+
          f'hidden layer config = {str(N_HIDDEN)}\n'+
          'solver = Adams')
    
    for t in sorted_list:
        f_log.write(f'{t[0]}    {t[1]*100:.2f}%    {t[2]}\n')

    f_log.write(f'\nBEST_REPEAT = {i_best}, with lowest MAPE = {mape_best*100:.2f}%')
    print(f'\nBEST_REPEAT = {i_best}, with lowest MAPE = {mape_best*100:.2f}%')
```

### 4. Pick and load the best DimNet model

* Prior to run the cell below, you need to manually type the BEST_REPEAT.
* If you've run the training cell more than once, you may compare different runs and pick the very best BEST_REPEAT
* more info can be found in log.txt the {NAME} directory. In this case, the path is smooth/log.txt

```python
BEST_REPEAT = 1

# load the correpsonding .model file corresponding to the best REPEAT #
from joblib import load

path = f'{NAME}/{NAME}_{BEST_REPEAT}.model'

model = load(path)
```

### 6. Evaluate the model performance

```python
y_pred = model.predict(df_train[col_x].values)
y_label = df_train.f.values

from core import compute_metrics
mape, p30, p50 = compute_metrics(y_label,y_pred, verbose=False)

fig, ax = plt.subplots(figsize=[4,4], dpi=100)

ax.plot(y_label, y_pred, marker='o', markersize=0.5, ls='')

ax.annotate(f"MAPE={mape*100:.1f}%\nP₃₀={p30*100:.1f}%\n", 
            xy=(0.05,0.95), xycoords='axes fraction', 
            horizontalalignment='left', verticalalignment='top', 
            fontsize='medium')

x = np.linspace(0,1,100)
lw = 0.75
ax.plot(x,x,'-.',c='k',linewidth=lw)
ax.plot(x,1.3*x,'--',c='k',linewidth=lw)
ax.plot(x,0.7*x,'--',c='k',linewidth=lw)
ax.annotate('+30%', xy=(0.7,0.95), xycoords='axes fraction',
            horizontalalignment='left', verticalalignment='top', fontsize='medium')
ax.annotate('-30%', xy=(0.88,0.78), xycoords='axes fraction', 
            horizontalalignment='left', verticalalignment='top', fontsize='medium')

ax.set_xlabel('label')
ax.set_ylabel('prediction')
ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlim(0.005,0.1)
ax.set_ylim(0.005,0.1)

plt.show()
```
